### Setting up the variables.

```python
X = df.iloc[:, [2,3]].values
y = df.iloc[:,-1].values
```

### Train Test Split.


```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 0)
```

### Scaling the data.

```python
from sklearn.preprocessing import StandardScaler as SScale
sc = SScale()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

### Creating a classifier object.

- DecisionTreeClassifier has criterion = 'gini' as default.
- criterion: the function to measure the quality of the split.
- criterion = 'gini' for gini impurity.
- criterion = 'entropy' for information gain.

```python
from sklearn.tree import DecisionTreeClassifier as DTC
classifier = DTC(criterion = 'entropy' , random_state = 0)
classifier.fit(X_train, y_train)
```

### Making predictions.

```python
predictions = classifier.predict(X_test)
predictions
```

### Confusion Matrix.

```python
from sklearn.metrics import confusion_matrix as CM

cm = CM(y_test, predictions)
cm
```

```python
plt.figure(figsize = (5,2))
sns.set(style = 'whitegrid', font_scale = 1.2) 

sns.heatmap(cm, annot = True, fmt = 'd', cmap ='coolwarm')
plt.show()
```

##### The classifier made 9 incorrect predictions.

### Classification Report.

```python
from sklearn.metrics import classification_report as CR
print(CR(y_test,predictions))
```
